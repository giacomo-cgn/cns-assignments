## Read PyTorch save files (extension ".pt")

```python
import torch
import numpy as np
```

```python
np.set_printoptions(threshold=10000) # Set print options to show full tensor
```

```python
pt_file_to_read = "LAB3_1/rnn_retrain_checkpoints/rnn_19.pt" # Use path
```

```python
def print_pytorch_dict(pytorch_dict):
    for k, v in pytorch_dict.items():
        if isinstance(v, dict):
            print(k)
            print_pytorch_dict(v)
        elif isinstance(v, torch.Tensor):
            print(k, v.shape)
            print(v)
        else:
            print(k)
            print(v)
        print('\n')
```

```python
# Read pytorch save file .pt

# Load the .pt file using PyTorch
loaded_data = torch.load(pt_file_to_read, map_location=torch.device('cpu'))  # Load data to CPU

print_pytorch_dict(loaded_data)


    
```
